```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
data = pd.read_csv("Mall_Customers.csv")
data.columns = ["CustomerID", "Genre", "Age", "Annual_income", "Spending_score"]
data.head()

# Any results you write to the current directory are saved as output.
```

```python
total_male_customers = data.groupby('Genre').get_group('Male').count()['CustomerID']
total_female_customers = data.groupby('Genre').get_group('Female').count()['CustomerID']
bar_plot_df = pd.DataFrame({'gender':['Male', 'Female'], 'frequency':[total_male_customers,total_female_customers]})
bar_plot_df.plot.bar(x='gender', y='frequency', rot=0)
```

```python
total_customers_by_age = data.groupby('Age').count().reset_index()
total_customers_by_age.sort_values('Age', ascending=False)
total_customers_by_age
```

```python
list_of_total_customers_by_age = total_customers_by_age['Age'].tolist()
list_of_freq_of_customers_by_age = total_customers_by_age['CustomerID'].tolist()
plt.bar(list_of_total_customers_by_age, list_of_freq_of_customers_by_age)
plt.show()
```

```python
avg_salaries_by_age = data.groupby('Age')['Annual_income'].mean().mul(1000).round(1)
plt.plot(list_of_total_customers_by_age,avg_salaries_by_age.tolist())
```

```python
avg_spending_scores_by_age = data.groupby('Age')['Spending_score'].mean().round(1)
plt.plot(list_of_total_customers_by_age,avg_spending_scores_by_age.tolist())
```
